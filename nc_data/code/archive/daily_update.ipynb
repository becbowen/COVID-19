```python
import pandas as pd
import numpy as np
import datetime as dt
import os
```

```python
#update path to covid daily update on sharepoint site
new_data = pd.read_csv("../../../databases/JHU/COVID-19/csse_covid_19_data/csse_covid_19_daily_reports/05-08-2020.csv")
new_data.head()
```

```python
new_data_nc = new_data[new_data["Province_State"] == "North Carolina" ]
new_data.head()
```

```python
new_data_nc = new_data_nc[["FIPS","Admin2","Province_State","Lat", "Long_", "Confirmed", "Last_Update"]]
new_data_nc.head()
```

```python
new_data_nc["Last_Update"] = pd.to_datetime(new_data_nc.Last_Update) 
new_data_nc.dtypes
```

```python
new_data_nc["Date"] = new_data_nc["Last_Update"].dt.to_period("D")
```

```python
offset = dt.timedelta(days = 1)
new_data_nc["Date"] = new_data_nc["Date"].apply(lambda x: x-offset)
new_data_nc.drop("Last_Update", axis = 1, inplace = True)
```

```python
new_data_nc.head()
```

```python
nc_time_series = pd.read_excel("../clean_data/nc_time_series.xlsx", index_col = 0)
```

```python
nc_time_series.tail()
```

```python
nc_time_series = nc_time_series.append(new_data_nc)
nc_time_series.tail()
```

```python
nc_time_series.reset_index(inplace = True)
nc_time_series.tail()
```

```python
nc_time_series.drop("index", axis = 1, inplace = True)
nc_time_series.head()
```

```python
nc_time_series.to_excel("../clean_data/nc_time_series.xlsx")
```
