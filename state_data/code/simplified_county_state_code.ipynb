```python
import pandas as pd
import numpy as np
import datetime 
import os 
```

```python
#read in the latest dataset
time_series = pd.read_csv("../../../databases/JHU/COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv")
time_series.head()
```

```python
#keep only rows with valid state and county data
time_series = time_series[time_series["Admin2"] != "Unassigned"]
time_series = time_series.dropna(subset = ["Admin2"])
time_series.head()
```

```python
time_series = time_series[~time_series["Admin2"].str.contains("Out of ")]
time_series.head()
```

```python
#define a dataframe with just the county information
all_counties = time_series[["FIPS", "Admin2","Province_State","Lat","Long_"]]
all_counties.head()
```

```python
#define a variable to define the end date of data to be included
startdate = datetime.datetime(2020,3,4)
today = datetime.datetime.today()
offset = 53 + (today-startdate).days
offset
```

```python
#define a dataframe with just the total cases by county each day
dates = time_series.iloc[:,53:offset]
dates.head()
```

```python
#define an empty dataframe to transpose the cases by date into a new row for each county-date pair
all_county_time_series = pd.DataFrame(columns = ["FIPS","Admin2","Province_State","Lat","Long_","Confirmed","Date"])
```

```python
#fill in the dataframe
cols = dates.columns

for i in range(0,len(cols)):
    one_day = pd.DataFrame(columns = ["Confirmed", "Date"])
    one_day["Confirmed"] = dates.iloc[:,i]
    one_day["Date"] = cols[i]
    county_time = all_counties.join(one_day)
    all_county_time_series = all_county_time_series.append(county_time)
#    county_time.append(one_day)
#    print(one_day)
all_county_time_series.head()
```

```python
all_county_time_series.reset_index(inplace = True)
all_county_time_series.drop("index", axis = 1, inplace = True)
```

```python
all_county_time_series["Date"] = pd.to_datetime(all_county_time_series.Date)
all_county_time_series.dtypes
```

```python
#convert Date column to format that will be uniform across use cases
#all_county_time_series["Date"] = all_county_time_series["Date"].dt.to_period("D")
```

```python
#add a column with a unique value that can be used to establish a relationship with other datasheets
all_county_time_series["Unique"] = all_county_time_series["Admin2"]+", "+all_county_time_series["Province_State"]+", "+all_county_time_series["Date"].astype(str)

#add a column so location will be readable by power BI
all_county_time_series["Location"] = all_county_time_series["Admin2"]+" County "+all_county_time_series["Province_State"]
```

```python
all_county_time_series.head()
```

```python
all_county_time_series.sort_values(by = ["Unique","Date"], inplace = True, ignore_index = True)
all_county_time_series.head()
```

```python
all_county_time_series["New_Cases"] = all_county_time_series["Confirmed"]-all_county_time_series["Confirmed"].shift(1)
all_county_time_series["Rolling_Avg"] = (all_county_time_series["Confirmed"]+all_county_time_series["Confirmed"].shift(1)+all_county_time_series["Confirmed"].shift(2)+all_county_time_series["Confirmed"].shift(3)+all_county_time_series["Confirmed"].shift(4)+all_county_time_series["Confirmed"].shift(5)+all_county_time_series["Confirmed"].shift(6)+all_county_time_series["Confirmed"].shift(7)+all_county_time_series["Confirmed"].shift(8)+all_county_time_series["Confirmed"].shift(9)+all_county_time_series["Confirmed"].shift(10)+all_county_time_series["Confirmed"].shift(11)+all_county_time_series["Confirmed"].shift(12)+all_county_time_series["Confirmed"].shift(13))/14
all_county_time_series["New_Case_Roll_Avg"] = (all_county_time_series["New_Cases"]+all_county_time_series["New_Cases"].shift(1)+all_county_time_series["New_Cases"].shift(2)+all_county_time_series["New_Cases"].shift(3)+all_county_time_series["New_Cases"].shift(4)+all_county_time_series["New_Cases"].shift(5)+all_county_time_series["New_Cases"].shift(6)+all_county_time_series["New_Cases"].shift(7)+all_county_time_series["New_Cases"].shift(8)+all_county_time_series["New_Cases"].shift(9)+all_county_time_series["New_Cases"].shift(10)+all_county_time_series["New_Cases"].shift(11)+all_county_time_series["New_Cases"].shift(12)+all_county_time_series["New_Cases"].shift(13))/14
all_county_time_series.head()
```

```python
timeframe = 14
all_county_time_series["New_Case_Change"] = (all_county_time_series["New_Case_Roll_Avg"]-all_county_time_series["New_Case_Roll_Avg"].shift(14))/timeframe
all_county_time_series["Normalized_Change"] = all_county_time_series.apply(lambda row: row.New_Case_Change/row.New_Case_Roll_Avg if row.New_Case_Roll_Avg !=0 else 0, axis = 1)
all_county_time_series.tail()
```

```python
offset2 = datetime.timedelta(days = 14)
firstdate = startdate + offset2 
firstdate
```

```python
all_county_time_series = all_county_time_series[all_county_time_series["Date"]>firstdate]
all_county_time_series.head()
```

```python
offset3 = datetime.timedelta(days = 2)
olddate = today - offset3
olddate
```

```python
all_county_time_series_today = all_county_time_series[all_county_time_series["Date"] > olddate]
```

```python
all_county_time_series = all_county_time_series[["Admin2","Province_State","Location","Unique","Date","Lat","Long_","Confirmed","Rolling_Avg",
                                "New_Cases","New_Case_Roll_Avg","New_Case_Change","Normalized_Change"]]
all_county_time_series_today = all_county_time_series_today[["Admin2","Province_State","Location","Unique","Date","Lat","Long_","Confirmed","Rolling_Avg",
                                "New_Cases","New_Case_Roll_Avg","New_Case_Change","Normalized_Change"]]
```

```python
#now collect the data by state
state_time_series = all_county_time_series.groupby(["Date","Province_State"])[["Confirmed"]].sum().reset_index()
state_time_series.head()
```

```python
state_time_series["Unique"] = state_time_series["Province_State"]+", "+state_time_series["Date"].astype(str)
state_time_series.tail()
```

```python
state_time_series.sort_values(by = ["Unique","Date"], inplace = True, ignore_index = True)
state_time_series["New_Cases"] = state_time_series["Confirmed"]-state_time_series["Confirmed"].shift(1)
state_time_series["Rolling_Avg"] = (state_time_series["Confirmed"]+state_time_series["Confirmed"].shift(1)+state_time_series["Confirmed"].shift(2)+state_time_series["Confirmed"].shift(3)+state_time_series["Confirmed"].shift(4)+state_time_series["Confirmed"].shift(5)+state_time_series["Confirmed"].shift(6)+state_time_series["Confirmed"].shift(7)+state_time_series["Confirmed"].shift(8)+state_time_series["Confirmed"].shift(9)+state_time_series["Confirmed"].shift(10)+state_time_series["Confirmed"].shift(11)+state_time_series["Confirmed"].shift(12)+state_time_series["Confirmed"].shift(13))/14
state_time_series["New_Case_Roll_Avg"] = (state_time_series["New_Cases"]+state_time_series["New_Cases"].shift(1)+state_time_series["New_Cases"].shift(2)+state_time_series["New_Cases"].shift(3)+state_time_series["New_Cases"].shift(4)+state_time_series["New_Cases"].shift(5)+state_time_series["New_Cases"].shift(6)+state_time_series["New_Cases"].shift(7)+state_time_series["New_Cases"].shift(8)+state_time_series["New_Cases"].shift(9)+state_time_series["New_Cases"].shift(10)+state_time_series["New_Cases"].shift(11)+state_time_series["New_Cases"].shift(12)+state_time_series["New_Cases"].shift(13))/14
state_time_series.head()
```

```python
timeframe = 14
state_time_series["New_Case_Change"] = (state_time_series["New_Case_Roll_Avg"]-state_time_series["New_Case_Roll_Avg"].shift(14))/timeframe
state_time_series["Normalized_Change"] = state_time_series.apply(lambda row: row.New_Case_Change/row.New_Case_Roll_Avg if row.New_Case_Roll_Avg !=0 else 0, axis = 1)
state_time_series.tail()
```

```python
offset2 = datetime.timedelta(days = 14)
firstdate = startdate + offset2 
firstdate
```

```python
state_time_series = state_time_series[state_time_series["Date"]>firstdate]
state_time_series.head()
```

```python
offset3 = datetime.timedelta(days = 2)
olddate = today - offset3
olddate
```

```python
state_time_series_today = state_time_series[state_time_series["Date"] > olddate]
```

```python
state_time_series = state_time_series[["Province_State","Unique","Date","Confirmed","Rolling_Avg",
                                "New_Cases","New_Case_Roll_Avg","New_Case_Change","Normalized_Change"]]
state_time_series_today = state_time_series_today[["Province_State","Unique","Date","Confirmed","Rolling_Avg",
                                "New_Cases","New_Case_Roll_Avg","New_Case_Change","Normalized_Change"]]
```
