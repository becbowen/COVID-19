```python
import pandas as pd
import numpy as np
import datetime
import os 
```

```python
rolling_avg = pd.read_excel("../clean_data/state_time_series.xlsx", index_col = 0)
rolling_avg.head()
```

```python
rolling_avg.sort_values(["Province_State", "Date"], inplace = True)
rolling_avg.reset_index(inplace = True)
rolling_avg.drop("index", axis = 1, inplace = True)
rolling_avg.head()
```

```python
rolling_avg["New_Cases"] = rolling_avg["Confirmed"]-rolling_avg["Confirmed"].shift(1)
rolling_avg = rolling_avg[rolling_avg["Date"] != "2020-03-18"]
rolling_avg.tail()
rolling_avg.tail()
```

```python
rolling_avg["Rolling_Avg"] = (rolling_avg["Confirmed"]+rolling_avg["Confirmed"].shift(1)+rolling_avg["Confirmed"].shift(2)+rolling_avg["Confirmed"].shift(3)+rolling_avg["Confirmed"].shift(4)+rolling_avg["Confirmed"].shift(5)+rolling_avg["Confirmed"].shift(6)+rolling_avg["Confirmed"].shift(7)+rolling_avg["Confirmed"].shift(8)+rolling_avg["Confirmed"].shift(9)+rolling_avg["Confirmed"].shift(10)+rolling_avg["Confirmed"].shift(11)+rolling_avg["Confirmed"].shift(12)+rolling_avg["Confirmed"].shift(13))/14
rolling_avg.head()
```

```python
rolling_avg["New_Case_Roll_Avg"] = (rolling_avg["New_Cases"]+rolling_avg["New_Cases"].shift(1)+rolling_avg["New_Cases"].shift(2)+rolling_avg["New_Cases"].shift(3)+rolling_avg["New_Cases"].shift(4)+rolling_avg["New_Cases"].shift(5)+rolling_avg["New_Cases"].shift(6)+rolling_avg["New_Cases"].shift(7)+rolling_avg["New_Cases"].shift(8)+rolling_avg["New_Cases"].shift(9)+rolling_avg["New_Cases"].shift(10)+rolling_avg["New_Cases"].shift(11)+rolling_avg["New_Cases"].shift(12)+rolling_avg["New_Cases"].shift(13))/14
rolling_avg.head()
```

```python
rolling_avg.tail()
```

```python
rolling_avg_change = rolling_avg[["Date","Province_State","Unique","Confirmed","Rolling_Avg","New_Cases","New_Case_Roll_Avg"]]
```

```python
timeframe = 14
rolling_avg_change["Change_New_Case_Avg"] = rolling_avg_change["New_Case_Roll_Avg"] - rolling_avg_change["New_Case_Roll_Avg"].shift(timeframe)
rolling_avg_change.tail(timeframe)
```

```python
rolling_avg_change["Normalized_Change"] = rolling_avg_change.apply(lambda row: row.Change_New_Case_Avg/row.New_Case_Roll_Avg if row.New_Case_Roll_Avg != 0 else 0, axis = 1)
rolling_avg_change.tail()
```

```python
today = datetime.datetime.today()
offset2 = datetime.timedelta(days = 2)
olddate = today - offset2
olddate
```

```python
rolling_avg_change["Date"] = pd.to_datetime(rolling_avg_change.Date)
rolling_avg_change_day = rolling_avg_change[rolling_avg_change["Date"] > olddate]

rolling_avg_change_day.tail()
```

```python
rolling_avg_change["Date"] = rolling_avg_change["Date"].dt.to_period("D")
rolling_avg_change_day["Date"] = rolling_avg_change_day["Date"].dt.to_period("D")
```

```python
rolling_avg_change.to_excel("../clean_data/state_rolling_avg.xlsx")
rolling_avg_change_day.to_excel("../clean_data/state_today.xlsx")
```
