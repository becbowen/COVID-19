```python
import pandas as pd
import numpy as np
import datetime
import os 
```

```python
time_series = pd.read_csv("../../databases/JHU/COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_US.csv")
time_series.head()
```

```python
time_series = time_series.dropna(subset = ["Admin2"])
time_series.head()
```

```python
sites = pd.read_excel("../clean_data/sites.xlsx")
sites.head()
```

```python
time_series_sites = time_series.merge(sites, left_on = "Combined_Key", right_on = "Combined_Key", how = "inner")
time_series_sites.head()
```

```python
counties_sites = time_series_sites[["FIPS", "Site", "Region", "County","State","Combined_Key"]]
counties_sites.head()
```

```python
startdate = datetime.datetime(2020,3,4)
today = datetime.datetime.today()
offset = 53 + (today-startdate).days
offset
```

```python
dates = time_series_sites.iloc[:,53:offset]
dates.head()
```

```python
county_time_series = pd.DataFrame(columns = ["FIPS","Site","Region","County","State","Combined_Key","Confirmed", 
                                             "Yesterday", "Day_3","Day_4","Day_5","Day_6","Day_7",
                                             "Day_8","Day_9","Day_10","Day_11","Day_12","Day_13","Day_14","Date"])
```

```python
#county_time = pd.DataFrame(columns = ["Confirmed"])
cols = dates.columns

for i in range(14,len(cols)):
    one_day = pd.DataFrame(columns = ["Confirmed", "Yesterday", "Day_3","Day_4","Day_5","Day_6","Day_7",
                                             "Day_8","Day_9","Day_10","Day_11","Day_12","Day_13","Day_14","Date"])
    one_day["Confirmed"] = dates.iloc[:,i]
    one_day["Yesterday"] = dates.iloc[:,i-1]
    one_day["Day_3"] = dates.iloc[:,i-2]
    one_day["Day_4"] = dates.iloc[:,i-3]
    one_day["Day_5"] = dates.iloc[:,i-4]
    one_day["Day_6"] = dates.iloc[:,i-5]
    one_day["Day_7"] = dates.iloc[:,i-6]
    one_day["Day_8"] = dates.iloc[:,i-7]
    one_day["Day_9"] = dates.iloc[:,i-8]
    one_day["Day_10"] = dates.iloc[:,i-9]
    one_day["Day_11"] = dates.iloc[:,i-10]
    one_day["Day_12"] = dates.iloc[:,i-11]
    one_day["Day_13"] = dates.iloc[:,i-12]
    one_day["Day_14"] = dates.iloc[:,i-13]
    one_day["Date"] = cols[i]
    county_time = counties_sites.join(one_day)
    county_time_series = county_time_series.append(county_time)
#    county_time.append(one_day)
#    print(one_day)
county_time_series.head()
```

```python
county_time_series.reset_index(inplace = True)
county_time_series.drop("index", axis = 1, inplace = True)
```

```python
county_time_series["Date"] = pd.to_datetime(county_time_series.Date)
county_time_series.dtypes
```

```python
county_time_series["Date"] = county_time_series["Date"].dt.to_period("D")
```

```python
county_time_series.tail()
```

```python
county_time_series["New_Cases"] = county_time_series.apply(lambda row: row.Confirmed - row.Yesterday, axis=1)
```

```python
county_time_series["Rolling_Avg"] = county_time_series.apply(
    lambda row: (row.Confirmed + row.Yesterday + row.Day_3+row.Day_4
                 +row.Day_5+row.Day_6+row.Day_7+row.Day_8+row.Day_9+
                 row.Day_10+row.Day_11+row.Day_12+row.Day_13+row.Day_14)/14, axis = 1)
```

```python
county_time_series = county_time_series[["FIPS", "Site", "Region", "County", "State", "Combined_Key", "Confirmed", 
                                         "New_Cases", "Rolling_Avg", "Date"]]
county_time_series.tail()
```

```python
county_time_series.sort_values(["Combined_Key", "Date"], inplace = True)
county_time_series.reset_index(inplace = True)
county_time_series.drop("index", axis = 1, inplace = True)
county_time_series.head(10)
```

```python
county_time_series.tail()
```

```python
county_time_series["Temp"] = county_time_series.Rolling_Avg.shift(1)
county_time_series.head()
```

```python
county_time_series["New_Case_Roll_Avg"] = county_time_series.apply(lambda row: row.Rolling_Avg - row.Temp, axis=1)
county_time_series = county_time_series[county_time_series["Date"] != "2020-03-18"]
county_time_series.tail()
```

```python
county_time_series = county_time_series[["FIPS","Site","Region","County","State","Combined_Key","Confirmed","Rolling_Avg",
                                       "New_Cases","New_Case_Roll_Avg","Date"]]
county_time_series.tail()
```

```python
county_time_series.to_excel("../clean_data/site_rolling_avg.xlsx")
```
